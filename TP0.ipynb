```python
from pyparsing.helpers import White
def create_russia_flag(width, height):
    img = Image.new('RGB', (width, height), 'white')
    draw = ImageDraw.Draw(img)
    white_height = height // 3
    draw.rectangle([0,0,  width, white_height ], fill='white')
    red_height = height - white_height
    draw.rectangle([0, white_height, width, white_height*2], fill=(0, 56, 168))
    draw.rectangle([0, white_height * 2, width, height], fill=(206, 17, 38))
    draw.rectangle([0, 0, width - 1, height - 1], outline='black')
    return img

 # Cr√©ez les images et enregistrez-les
russia_flag = create_russia_flag(300, 200)
russia_flag.save('/content/russia_flag.png')
russia_flag

```
